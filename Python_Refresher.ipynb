```python
# Ask user for input - number of days
days = int(input("Enter the number of days: "))

# Calculate number of seconds
# 1 day = 24 hours, 1 hour = 60 minutes, 1 minute = 60 seconds
seconds = days * 24 * 60 * 60

# Display the result
print(f"{days} days is equal to {seconds:,} seconds")
```

```python
#program to calculate volume of a sphere
import math

#input radius of the sphere
radius = float(input("Enter the radius of the sphere: "))

#calculate the volume of the sphere using the exponential operator
volume = (4/3) * math.pi * (radius ** 3)

#output
print(f"The volume of the sphere with raduis {radius} is {volume: 2f}")

```

```python
#program for area and perimeter of a square using functions
def area(side):
  return side*side

def perimeter(side):
  return 4 * side
#get side length from user
side_length = int(input("Enter the side lenth of the square"))

#calculate area and length
area = area(side_length)
perimeter = perimeter(side_length)

#output
print(f"A square with side length of {side_length} has an area of {area} and perimeter of {perimeter}")
```

```python
#program that determines lowercase and upercase using functions
def case(char):
  if char.isupper():
    return "uppercase"
  elif char.islower():
    return "lowercase"
  else:
    return "not categorized"

#get input from user
char = input("Enter any character")

#call the function to store the result
result = case(char)

#output
print(f"The character {char} is {result}")
```

```python
#initialize variables
x = 0
y = 20

#REPEAT-UNTIL loop
while True:
    y -= 4
    #add 2/y to x
    if y != 0:
        x += 2 / y

    #UNTIL condition
    if y < 6:
        break

#output x
print(f"The final value of x is: {x}")
```

```python
#program to input 5 values and calculate their average

values = []

#use a loop to continually input 5 values
for i in range(5):
      num = float(input(f"Enter value {i + 1} of 5: "))
      values.append(num)

#calculate the average of the values
average = sum(values) / len(values)

#output
print(f"\nThe values you entered are: {values}")
print(f"The average of these values is: {average}")
```
